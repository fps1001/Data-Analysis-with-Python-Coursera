Handle missing data:

- For continuous attributes (e.g., age), replace missing values with the mean.
- For categorical attributes (e.g., smoker), replace missing values with the most frequent value.
- Update the data types of the respective columns.
- Verify the update using `df.info()`.


````python
# Age Dtype to int and missing values to mean.
# Ensure age is numeric first (coerce invalid entries to NaN), then compute mean and fill NaNs.
data['age'] = pd.to_numeric(data['age'], errors='coerce') # coerce means invalid parsing will be set as NaN
mean_age = data['age'].mean()
# Filling NaN values with mean age
data['age'] = data['age'].fillna(mean_age)
# Converting age to int
data['age'] = data['age'].round().astype(int)
````

````python
# Smoker to int and missing values to most frequent value.
most_frequent_smoker = data['smoker'].value_counts().idxmax()
# Avoid pandas FutureWarning from inplace on a Series slice by assigning the filled Series back to the DataFrame
data['smoker'] = data['smoker'].fillna(most_frequent_smoker)
# Now convert smoker column to integer (safe because NaNs were filled)
data['smoker'] = data['smoker'].astype(int)

data.info()
````

Create a training pipeline that uses `StandardScaler()`, `PolynomialFeatures()` and `LinearRegression()` to create a model that can predict the `charges` value using all the other attributes of the dataset. There should be even further improvement in the performance.

Apply polynomial transformation to the training parameters with degree=2. Use this transformed feature set to fit the same regression model, as above, using the training subset. Print the $ R^2 $ score for the testing subset.


````python
# Polinomial transformation with degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Split the polynomial features into training and testing sets
X_poly_train, X_poly_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)

ridge = Ridge(alpha=0.1)
# Fit the model with the ridge regression
ridge.fit(X_poly_train, y_train)

# Check the score on the test set
print(f'Ridge Regression model score on test set: {ridge.score(X_poly_test, y_test)}')
````
